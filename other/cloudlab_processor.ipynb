```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
link_cloud_lab_processor = "https://docs.cloudlab.us/hardware.html"
reqs = requests.get(link_cloud_lab_processor)
soup = BeautifulSoup(reqs.text, 'html.parser')
```

```python
# Find all the tables with class="boxed"
all_tables = soup.find_all('table', class_='boxed')
```

```python
all_tables[0]
```

```python
df = pd.read_html(str(all_tables))
```

```python
len(df)
```

```python
df_cloudlab_processor = pd.DataFrame()
```

```python
# Iterate over each table and convert it to a DataFrame
for table in all_tables:
    # Convert the table to a DataFrame
    df = pd.read_html(str(table))[0]
    
    # Rename the 'A' column to 'X'
  
    df_cloudlab_processor = pd.concat([df_cloudlab_processor, df], ignore_index=False)
```

```python
df_cloudlab_processor
```

```python
df_cloudlab_processor.columns
```

```python
df_cloudlab_processor = df_cloudlab_processor.drop(columns=[1])
```

```python
df_cloudlab_processor
```

```python
df_cloudlab_processor = df_cloudlab_processor.rename(columns={0: "resource", 2: "description"})
```

```python
df_cloudlab_processor
```

```python
df_cloudlab_processor_intel = df_cloudlab_processor[
    df_cloudlab_processor["description"].str.contains("Intel", regex=False, case=False)]
```

```python
df_cloudlab_processor_intel
```

```python
df_cloudlab_processor_intel = df_cloudlab_processor_intel[df_cloudlab_processor_intel["resource"]=="CPU"]
```
