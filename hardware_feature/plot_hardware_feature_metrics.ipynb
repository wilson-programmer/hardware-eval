```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_patch_stat = pd.read_csv("patch_feature.csv")
df_patch_stat
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Données exemple
categories = [feature.upper() for feature in df_patch_stat["feature"]]
data = [df_patch_stat["added_lines"],
        df_patch_stat["deleted_lines"],
        df_patch_stat["updated_files"],
        df_patch_stat["#patch"]]

# Création du barplot
x = np.arange(len(categories))  # Positions des barres sur l'axe x
width = 0.23  # Largeur des barres

fig, ax = plt.subplots(figsize=(14, 6))
rects1 = ax.bar(x - width, data[0], width, label='Added LOCs')
rects2 = ax.bar(x, data[1], width, label='Removed LOCs') 
rects3 = ax.bar(x + width, data[2], width, label='Files changed')
rects4 = ax.bar(x + width*2, data[3], width, label='Number of patchs') 

# Affichage des valeurs au-dessus des barres
for rects in [rects1, rects2, rects3, rects4]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Ajout d'un titre et des étiquettes des axes
#ax.set_title('Barplot avec plusieurs catégories et valeurs')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_xlabel('Hardware features')
#ax.set_ylabel('Valeurs')
ax.legend()

path= "hardware_feature/"
#Export to pdf format
plt.savefig("hardware_feature_stat.pdf", format="pdf", bbox_inches="tight")

#Export to svg format
plt.savefig("hardware_feature_stat.png", format="png", bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Données exemple
categories = ['Catégorie A', 'Catégorie B', 'Catégorie C']
data = [[10, 15, 20], 
        [12, 18, 25],
        [8, 13, 18]]

# Création du barplot
x = np.arange(len(categories))  # Positions des barres sur l'axe x
width = 0.2  # Largeur des barres

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, data[0], width, label='Série 1')
rects2 = ax.bar(x, data[1], width, label='Série 2') 
rects3 = ax.bar(x + width, data[2], width, label='Série 3')

# Affichage des valeurs au-dessus des barres
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Ajout d'un titre et des étiquettes des axes
ax.set_title('Barplot avec plusieurs catégories et valeurs')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_xlabel('Catégories')
ax.set_ylabel('Valeurs')
ax.legend()

plt.show()
```
