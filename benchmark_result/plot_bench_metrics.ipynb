```python
#sequential read throughput
mean_seq_read_throughput2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "seq_read_throughput_round",
                            regex=False, case=False)]["Throughput"])

mean_seq_read_throughput_mba2= np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "seq_read_throughput_mba_round",
                            regex=False, case=False)]["Throughput"])

#sequential write throughput
mean_seq_write_throughput2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "seq_write_throughput_round",
                            regex=False, case=False)]["Throughput"])

mean_seq_write_throughput_mba2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "seq_write_throughput_mba_round",
                            regex=False, case=False)]["Throughput"])

#random read throughput
mean_rand_read_throughput2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "rand_read_throughput_round",
                            regex=False, case=False)]["Throughput"])

mean_rand_read_throughput_mba2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "rand_read_throughput_mba_round",
                            regex=False, case=False)]["Throughput"])

#random write throughput
mean_rand_write_throughput2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "rand_write_throughput_round",
                            regex=False, case=False)]["Throughput"])

#random write throughput
mean_rand_write_throughput_mba2 = np.mean(df_fio_bench_result[df_fio_bench_result["Op"].str.contains(
                            "rand_write_throughput_mba_round",
                            regex=False, case=False)]["Throughput"])
```

```python
# Catégories pour l'axe x
categories = ['seq_read', 'seq_write', 'rand_read', 'rand_write']

# Valeurs pour la première série de données
valeurs1 = [mean_seq_read_throughput2, mean_seq_write_throughput2, mean_rand_read_throughput2,
            mean_rand_write_throughput2]

# Valeurs pour la deuxième série de données
valeurs2 = [mean_seq_read_throughput_mba2, mean_seq_write_throughput_mba2, mean_rand_read_throughput_mba2,
            mean_rand_write_throughput_mba2]

# Largeur des barres
largeur_barre = 0.35
# Couleurs des barres
couleur1 = 'blue'
couleur2 = 'red'

# Position des barres sur l'axe x
pos = np.arange(len(categories))

# Créer le diagramme en barres
fig, ax = plt.subplots()
barre1 = ax.bar(pos, valeurs1, largeur_barre, color=couleur1, label='xen 4.10.4')
barre2 = ax.bar(pos + largeur_barre, valeurs2, largeur_barre, color=couleur2, label='xen 4.10.4 mba')

# Ajouter des étiquettes, une légende et un titre
ax.set_xlabel('Operations')
ax.set_ylabel('Bandwidth (MB/s)')
ax.set_title('Fio benchmark bandwidth')
ax.set_xticks(pos + largeur_barre / 2)
ax.set_xticklabels(categories)
ax.legend()


path= "fio/"
#Export to pdf format
plt.savefig(path+"fio_bandwidth.pdf", format="pdf", bbox_inches="tight")

#Export to svg format
plt.savefig(path+"fio_bandwidth.svg", format="svg", bbox_inches="tight")
plt.figure(figsize=(8,6))
# Afficher le diagramme en barres
plt.show()
```

```python
import re

def find_404_error(filename:str):
    """Check if the content of a patch is not correct
       
       Agrs:
           filename(str): the patch file to check
       Returns:
           
    """
    with open(filename) as f:
        contents = f.read()
  
    match = re.search("404 - Unknown commit object", contents)

    if match:
        return True
    else:
        return False

# Test
print(find_404_error("ecosystem_metrics/xen_4-10-4/d176cd6194.diff"))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['A', 'B', 'C', 'D']
data1 = [10, 15, 12, 18]
data2 = [8, 12, 10, 14]
data3 = [12, 14, 16, 13]
data4 = [9, 11, 13, 15]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, d1, d2, d3, d4 in zip(axes.flat, data1, data2, data3, data4):
    x = np.arange(len(labels))  # the label locations
    width = 0.15  # the width of the bars

    rects1 = ax.bar(x - 1.5*width, d1, width, label='Line 1')
    rects2 = ax.bar(x - 0.5*width, d2, width, label='Line 2')
    rects3 = ax.bar(x + 0.5*width, d3, width, label='Line 3')
    rects4 = ax.bar(x + 1.5*width, d4, width, label='Line 4')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Values')
    ax.set_title(f'Bar Plot with Four Lines')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

plt.suptitle('Multiple Bar Plots with Four Lines', fontsize=16)
plt.show()
```
