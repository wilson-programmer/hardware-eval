```python
# Catégories pour l'axe x
categories = ["PING_INLINE", "PING_MBULK", "SET", "GET"]

# Valeurs pour la première série de données
#valeurs1 = [853.7, 527.5]

# Valeurs pour la deuxième série de données
#valeurs2 = [863.0, 519.6,]


#Mean

valeurs1_mean_rps = [np.mean(set_rps_xen_4_10_4), np.mean(get_rps_xen_4_10_4), 
                 np.mean(ping_inline_rps_xen_4_10_4), np.mean(ping_mbulk_rps_xen_4_10_4)]
# Valeurs pour la deuxième série de données
valeurs2_mean_rps = [np.mean(set_rps_xen_4_11_1), np.mean(get_rps_xen_4_11_1), 
                 np.mean(ping_inline_rps_xen_4_11_1), np.mean(ping_mbulk_rps_xen_4_11_1)]

std1_rps = [np.std(set_rps_xen_4_10_4), np.std(get_rps_xen_4_10_4), 
                 np.std(ping_inline_rps_xen_4_10_4), np.std(ping_mbulk_rps_xen_4_10_4)]
# Valeurs pour la deuxième série de données
std2_rps = [np.std(set_rps_xen_4_11_1), np.std(get_rps_xen_4_11_1), 
                 np.std(ping_inline_rps_xen_4_11_1), np.std(ping_mbulk_rps_xen_4_11_1)]




#Median

valeurs1_med_rps = [np.median(set_rps_xen_4_10_4), np.median(get_rps_xen_4_10_4), 
                 np.median(ping_inline_rps_xen_4_10_4), np.median(ping_mbulk_rps_xen_4_10_4)]
# Valeurs pour la deuxième série de données
valeurs2_med_rps = [np.median(set_rps_xen_4_11_1), np.median(get_rps_xen_4_11_1), 
                 np.median(ping_inline_rps_xen_4_11_1), np.median(ping_mbulk_rps_xen_4_11_1)]

#Percentile 95th

valeurs1_perc95_rps = [np.percentile(set_rps_xen_4_10_4, 95), np.percentile(get_rps_xen_4_10_4, 95), 
                 np.percentile(ping_inline_rps_xen_4_10_4, 95), np.percentile(ping_mbulk_rps_xen_4_10_4, 95)]
# Valeurs pour la deuxième série de données
valeurs2_perc95_rps = [np.percentile(set_rps_xen_4_11_1, 95), np.percentile(get_rps_xen_4_11_1, 95), 
                 np.percentile(ping_inline_rps_xen_4_11_1, 95), np.percentile(ping_mbulk_rps_xen_4_11_1, 95)]





# Largeur des barres
largeur_barre = 0.35
# Couleurs des barres
couleur1 = 'blue'
couleur2 = 'red'

# Position des barres sur l'axe x
pos = np.arange(len(categories))

# Créer le diagramme en barres
#fig, (ax_mean, ax_med) = plt.subplots(1, 2, figsize=(12, 5.1))
fig, ((ax_mean, ax_med), (ax_perc95, ax_none)) = plt.subplots(2, 2, figsize=(12, 12))

#Mean

barre1_mean = ax_mean.bar(pos, valeurs1_mean_rps, largeur_barre, yerr=std1_rps, color=couleur1, 
                label='xen 4.10.4', capsize=3)
barre2_mean = ax_mean.bar(pos + largeur_barre, valeurs2_mean_rps, largeur_barre,yerr=std2_rps,
                color=couleur2, label='xen 4.11.1', capsize=3)

# Ajouter des étiquettes, une légende et un titre
ax_mean.set_xlabel('Test')
ax_mean.set_ylabel('Unit (rps)')
ax_mean.set_title('Mean')
ax_mean.set_xticks(pos + largeur_barre / 2)
ax_mean.set_xticklabels(categories)
ax_mean.legend(bbox_to_anchor=(0.5, 1), loc='upper left')


#Label the bar

for i in pos:
    if valeurs1_mean[i] < valeurs2_mean[i]:
        gain = (valeurs2_mean[i]-valeurs1_mean[i])/valeurs1_mean[i] * 100
        label = f'+{gain:.1f}%'
        ax_mean.text(i, valeurs2_mean[i]+0.1, label, color='g', rotation=45)
    else:
        gain = (valeurs1_mean[i]-valeurs2_mean[i])/valeurs2_mean[i] * 100
        label = f'+{gain:.1f}%'
        ax_mean.text(i, valeurs1_mean[i]+0.1, label, color='g', rotation=45)
        
        

#Median

barre1_med = ax_med.bar(pos, valeurs1_med_rps, largeur_barre, yerr=std1_rps, color=couleur1, 
                label='xen 4.10.4', capsize=3)
barre2_med = ax_med.bar(pos + largeur_barre, valeurs2_med_rps, largeur_barre,yerr=std2_rps,
                color=couleur2, label='xen 4.11.1', capsize=3)

# Ajouter des étiquettes, une légende et un titre
ax_med.set_xlabel('Test')
ax_med.set_ylabel('Unit (rps)')
ax_med.set_title('Median')
ax_med.set_xticks(pos + largeur_barre / 2)
ax_med.set_xticklabels(categories)
ax_med.legend(bbox_to_anchor=(0.5, 1), loc='upper left')


#Label the bar

for i in pos:
    if valeurs1_med_rps[i] < valeurs2_med_rps[i]:
        gain = (valeurs2_med_rps[i]-valeurs1_med_rps[i])/valeurs1_med_rps[i] * 100
        label = f'+{gain:.1f}%'
        ax_med.text(i, valeurs2_med_rps[i]+0.1, label, color='g', rotation=45)
    else:
        gain = (valeurs1_med_rps[i]-valeurs2_med_rps[i])/valeurs2_med_rps[i] * 100
        label = f'+{gain:.1f}%'
        ax_med.text(i, valeurs1_med_rps[i]+0.1, label, color='g', rotation=45)



#Percentile 95th

barre1_perc95 = ax_perc95.bar(pos, valeurs1_perc95_rps, largeur_barre, yerr=std1_rps, color=couleur1, 
                label='xen 4.10.4', capsize=3)
barre2_perc95 = ax_perc95.bar(pos + largeur_barre, valeurs2_perc95_rps, largeur_barre,yerr=std2_rps,
                color=couleur2, label='xen 4.11.1', capsize=3)

# Ajouter des étiquettes, une légende et un titre
ax_perc95.set_xlabel('Test')
ax_perc95.set_ylabel('Unit (rps)')
ax_perc95.set_title('Percentile 95th')
ax_perc95.set_xticks(pos + largeur_barre / 2)
ax_perc95.set_xticklabels(categories)
ax_perc95.legend(bbox_to_anchor=(0.5, 1), loc='upper left')


#Label the bar

for i in pos:
    if valeurs1_perc95_rps[i] < valeurs2_perc95_rps[i]:
        gain = (valeurs2_perc95_rps[i]-valeurs1_perc95_rps[i])/valeurs1_perc95_rps[i] * 100
        label = f'+{gain:.1f}%'
        ax_perc95.text(i, valeurs2_perc95_rps[i]+0.1, label, color='g', rotation=45)
    else:
        gain = (valeurs1_perc95_rps[i]-valeurs2_perc95_rps[i])/valeurs2_perc95_rps[i] * 100
        label = f'+{gain:.1f}%'
        ax_perc95.text(i, valeurs1_perc95_rps[i]+0.1, label, color='g', rotation=45)




# Center the second subplot
ax2_perc95_pos = ax_perc95.get_position()
ax_perc95.set_position([0.25, ax2_perc95_pos.y0, 0.5, ax2_perc95_pos.height])

# Hide the second subplot
ax_none.set_visible(False)        
        
        
        
        
fig.suptitle('Redis benchmark of 1 VM among 10 with 10 iterations', fontsize=14)




path= "redis/"
#Export to pdf format
plt.savefig(path+"redis_rps_vm0.pdf", format="pdf", bbox_inches="tight")

#Export to svg format
plt.savefig(path+"redis_rps_vm0.svg", format="svg", bbox_inches="tight")

#plt.figure(figsize=(8,6))
# Afficher le diagramme en barres
plt.show()
```
