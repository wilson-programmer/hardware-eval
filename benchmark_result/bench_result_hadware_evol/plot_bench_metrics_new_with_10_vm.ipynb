## Function to get statistical metrics for a specific xen version (FIO)

```python
def compute_stat_metrics(path_xen_version: str, metric_to_extract: str)->[list]:
    """Function to collect statistics metrics for a specific version of xen
        
       Args:
            path_xen_version (str) : Path of xen version to extract statistical metrics 
            metric_to_extract (str) : Metric to extract, e.g : Throughput or IOPS
            
        
        Returns:
            all_fio_metric ([list]) : A list containing all statistical matrics, each element is a list
    """
    
    #Read the file 
    df_fio_xen_version = pd.read_csv(path_xen_version)
    
    #Statistical metrics
    
    #Mean, Median and percentile 95th
    
    mesure_list = ["seq_read_throughput_round", "seq_write_throughput_round",
                  "rand_read_throughput_round", "rand_write_throughput_round"]
    
    mean_values = []
    med_values = []
    perc95_values = []
    errorbar_values = []
    
    for mesure in mesure_list:
        metric_mean = np.mean(df_fio_xen_version[df_fio_xen_version["Op"].str.contains(
                            mesure, regex=False, case=False)][metric_to_extract])
        
        metric_med = np.median(df_fio_xen_version[df_fio_xen_version["Op"].str.contains(
                            mesure, regex=False, case=False)][metric_to_extract])
        
        metric_perc95 = np.percentile(df_fio_xen_version[df_fio_xen_version["Op"].str.contains(
                            mesure, regex=False, case=False)][metric_to_extract], 95)
        
        
        metric_errorbar = np.std(df_fio_xen_version[df_fio_xen_version["Op"].str.contains(
                            mesure, regex=False, case=False)][metric_to_extract])
        
        mean_values.append(metric_mean)
        
        med_values.append(metric_med)
        
        perc95_values.append(metric_perc95)
        
        errorbar_values.append(metric_errorbar)
        
    all_fio_metric = [mean_values, med_values, perc95_values, errorbar_values]
    
    return all_fio_metric
    
    
```

## Function to get statistical metrics for a specific xen version (REDIS)

```python
def compute_stat_metrics_redis(path_xen_version: str, metric_to_extract: str)->[list]:
    """Function to collect statistics metrics for a specific version of xen for Redis
        
       Args:
            path_xen_version (str) : Path of xen version to extract statistical metrics 
            metric_to_extract (str) : Metric to extract, e.g : Latency or RPS
            
        
        Returns:
            all_fio_metric ([list]) : A list containing all statistical matrics, each element is a list
    """
    #import file
    redis_bench_xen_version = path_xen_version
    redis_bench_path_xen_version = glob(os.path.join(redis_bench_xen_version, "*.csv"))
    
    #Define the tests we are going to use
    #Iterate over each file and get the value of the define test

    set_ = []
    get_ = []
    ping_inline_ = []
    ping_mbulk_ = []

    
    for bench in redis_bench_path_xen_version:
        
        #Read a result file and extract metrics
        bench_result = pd.read_csv(bench)
        
        set_.append(((bench_result[bench_result["test"]=="SET"][metric_to_extract]).to_list())[0])
        get_.append(((bench_result[bench_result["test"]=="GET"][metric_to_extract]).to_list())[0])
        ping_inline_.append(((bench_result[bench_result["test"]=="PING_INLINE"][metric_to_extract]).to_list())[0])
        ping_mbulk_.append(((bench_result[bench_result["test"]=="PING_MBULK"][metric_to_extract]).to_list())[0])
        
    
    mean_values = [np.mean(set_), np.mean(get_), np.mean(ping_inline_), np.mean(ping_mbulk_)]
    
    med_values =  [np.median(set_), np.median(get_), np.median(ping_inline_), np.median(ping_mbulk_)]
    
    perc95_values = [np.percentile(set_, 95), np.percentile(get_, 95), 
                     np.percentile(ping_inline_, 95), np.percentile(ping_mbulk_, 95)]
    std_error =  [np.std(set_), np.std(get_), np.std(ping_inline_), np.std(ping_mbulk_)]
    
    all_metrics = [mean_values, med_values, perc95_values, std_error]
    
    return all_metrics
```

```python
def compute_stat_metrics_apache(path_xen_version: str)->[list]:
    """Function to collect statistics metrics for a specific version of xen for Apache
        
       Args:
            path_xen_version (str) : Path of xen version to extract statistical metrics 
            
            
        
        Returns:
            df_apache_metrics (DataFrame) : A DataFrame containing all statistical matrics.
    """
    #import file
    apache_bench_xen_version = path_xen_version
    apache_bench_path_xen_version = glob(os.path.join(apache_bench_xen_version, "*.csv"))
    
    #Define the tests we are going to use
    #Iterate over each file and get the value of the define test
    #list to contain all dataframe for all rounds
    mean_values = []
    med_values = []
    perc95_values = []
    std_error_values = []
    for i in range(101):
        value_of_each_round = []
        for bench in apache_bench_path_xen_version:       
            bench_result = pd.read_csv(bench)["Time in ms"][i]
            value_of_each_round.append(bench_result)

        mean_values.append(np.mean(value_of_each_round))
        med_values.append(np.median(value_of_each_round))
        perc95_values.append(np.percentile(value_of_each_round, 95))
        std_error_values.append(np.std(value_of_each_round))
        
    percentage_served = [i for i in range (101)]    
    df_apache_metrics = pd.DataFrame({"percentage_served": percentage_served, "mean":mean_values, 
                                      "median": med_values, "perc95":perc95_values, "error_bar": std_error_values})
    #all_metrics = [mean_values, med_values, perc95_values, std_error_values]
    return df_apache_metrics
```

```python
path_apache_bench_xen_4_10_4 = "apache/xen_4_10_4/vm0/apache_result"
path_apache_bench_xen_4_11_1 = "apache/xen_4_11_1/vm0/apache_result"
path_apache_bench_xen_4_11_2 = "apache/xen_4_11_2/vm0/apache_result"
path_apache_bench_xen_4_11_3 = "apache/xen_4_11_3/vm0/apache_result"


apache_bench_xen_4_10_4 = compute_stat_metrics_apache(path_apache_bench_xen_4_10_4)
apache_bench_xen_4_11_1 = compute_stat_metrics_apache(path_apache_bench_xen_4_11_1)
apache_bench_xen_4_11_2 = compute_stat_metrics_apache(path_apache_bench_xen_4_11_2)
apache_bench_xen_4_11_3 = compute_stat_metrics_apache(path_apache_bench_xen_4_11_3)
```
